# CFNXX — Skeptical-Reviewer Numerical Notebook Template

**Purpose.** This template is for a CFN notebook that sounds human on first contact and adversarial on second contact.  
Every markdown cell is split into two ordered parts:

1. **Friendly orientation** — plain-language setup in the style of CFN04.
2. **Mechanical review contract** — the exact burden, constraints, and notebook obligations.

### Friendly orientation

Use the first half of each markdown cell to tell the reader what the section is doing, why it matters, and what they should look for. It should read like CFN04: clear, concrete, and easy to follow without diluting the scientific burden.

### Mechanical review contract

## Template-wide markdown rule

- Every non-title markdown cell must appear in this order:
  1. `### Friendly orientation`
  2. `### Mechanical review contract`
- The friendly half explains the section in ordinary language.
- The mechanical half states the exact rules, burdens, and failure conditions.
- The friendly half must **not** replace the hard contract.
- The hard contract must **not** be softened into vague prose.

## Non-negotiable template rules

1. Put **only the paper's own explicit claims and paper-level gates** in the front matter.
2. Do **not** check whether the source paper file exists.
3. Do **not** read external CSV manifests or release contracts inside the notebook.
4. Do **not** open with internal audit gates like `T0`, `T1`, or `T2`.
5. Every claim section must attack the claim directly with:
   - the scientific burden,
   - what would kill the claim,
   - the attack design,
   - at least one wrong-form / negative control,
   - one decision figure,
   - one result block.
6. Paper-level gates should reuse prior evidence when possible instead of rerunning the same burden.
7. Put any publication-package note in a short appendix, **not** in the notebook front matter.


## 0. Setup

### Friendly orientation

This is the staging area. Load imports, fix the seed, and define small helpers so every later attack is reproducible. Nothing scientific happens here yet; the point is to make the notebook boring, stable, and easy to trust.

### Mechanical review contract

- Deterministic execution only.

```python

from __future__ import annotations

from dataclasses import dataclass
from typing import Any, Dict, List

import numpy as np
import matplotlib.pyplot as plt

SEED = 20260402
rng = np.random.default_rng(SEED)
plt.rcParams.update({"figure.dpi": 110, "figure.figsize": (8.0, 3.8), "axes.grid": True})

@dataclass
class ReviewEntry:
    unit_id: str
    title: str
    passed: bool
    metrics: Dict[str, Any]
    thresholds: Dict[str, Any]
    notes: str

LEDGER: List[ReviewEntry] = []
RESULTS: Dict[str, Dict[str, Any]] = {}

def terminal_log(unit_id: str, title: str, passed: bool, metrics: Dict[str, Any], criteria: Dict[str, Any], notes: str = "") -> None:
    status_symbol = "✅" if passed else "❌"
    verdict = "PASS" if passed else "FAIL"
    print("=" * 96)
    print(f"{unit_id} — {title}")
    print(f"Reviewer verdict: {status_symbol} {verdict}")
    print("Thresholds:")
    for k, v in criteria.items():
        print(f"  - {k}: {v}")
    print("Observed:")
    for k, v in metrics.items():
        print(f"  - {k}: {v}")
    if notes:
        print("Interpretation:")
        print(f"  {notes}")
    print("=" * 96)

def append_gate(unit_id: str, title: str, passed: bool, metrics: Dict[str, Any], criteria: Dict[str, Any], notes: str) -> None:
    LEDGER.append(ReviewEntry(unit_id, title, passed, metrics, criteria, notes))
    RESULTS[unit_id] = metrics

print({"seed": SEED, "standalone": True, "external_file_io": False})
```
